## 1. ממוצע
</div>


<div dir='rtl'>
הממוצע החשבוני (Average או Mean) מסומן על ידי 
</div>


$$\overline{x}$$

 <div dir='rtl'>
 ומוגדר על ידי חלוקה של סכום הנתונים במספר השכיחויות. <br>
אם נסמן את הנתונים בעזרת 
</div>

$$x_i$$

<div dir='rtl'>
עבור הערכים
</div>

 $$1\leq i\leq n$$

 <div dir='rtl'>
 (כאשר יש n נתונים), נקבל את הנוסחה: 
 </div>

$$ \overline{x} = \sum_{i=1}^{n} \frac{x_i}{n} = \frac{\sum_{i=1}^{n} x_i}{n} $$

<div dir='rtl'>
ניקח לדוגמה נתונים של 25,000 בני אדם בגיל 18, עם הגבהים והמשקלים שלהם.<br>
הנתונים נלקחו מהאתר:  http://socr.ucla.edu/docs/resources/SOCR_Data/SOCR_Data_Dinov_020108_HeightsWeights.html
</div>


```python
url = "http://socr.ucla.edu/docs/resources/SOCR_Data/SOCR_Data_Dinov_020108_HeightsWeights.html"
Heights_and_Weights_df = pd.read_html(url, header=0)[0]
Heights_and_Weights_df.set_index('Index', inplace=True)
Heights_and_Weights_df.rename_axis(None, axis=0, inplace=True)
```

<div dir='rtl'>
הנתונים המקוריים הם של גובה באינצ'ים ומשקל בפאונדים.<br>
הנתונים מרובים, חמשת הנתונים הראשונים וחמשת הנתונים האחרונים הם:
</div>


```python
Heights_and_Weights_df
```

<div dir='rtl'>
בשביל הנוחות, נשנה את הכותרות ואת הנתונים לסנטימטרים וקילוגרמים בהתאמה.
</div>


```python
Heights_and_Weights_df.columns = ['Height (CM)','Weight (KG)']
Heights_and_Weights_df["Height (CM)"] *= 2.54
Heights_and_Weights_df["Weight (KG)"] *= 0.453592
Heights_and_Weights_df
```

אם היינו סוכמים את כל המשקלים ומחלקים במספר הנתונים (25,000), היינו מקבלים את התוצאה הבאה:
</div>

```python
mean_weight = Heights_and_Weights_df['Weight (KG)'].mean()
mean_weight
```

<div dir='rtl'>

בצורה דומה עבור הגבהים:
</div>

```python
mean_height = Heights_and_Weights_df['Height (CM)'].mean()
mean_height
```

<div dir='rtl'>

בואו נראה את גרף הנתונים, ונוסיף נקודה אדומה למיקום שמתאים לממוצעים של הגבהים והמשקלים.
</div>

```python
plt.scatter(Heights_and_Weights_df["Height (CM)"], Heights_and_Weights_df["Weight (KG)"])
plt.scatter(mean_height, mean_weight, color='red')

plt.xlabel("Height (CM)")
plt.ylabel("Weight (KG)")

plt.show()
```

## חציון (Median)
</div>

<div dir='rtl'>
החציון הוא ערך החוצה את קבוצת הנתונים, עם מספר שווה של ערכים מעליו ומתחתיו.<br>
למשל החציון של הקבוצה 
</div>

$$\{1,2,10\}$$

<div dir='rtl'>
 הוא
</div>

$$Me=2$$

<div dir='rtl'>
כאשר מספר הנתונים אי-זוגי הגדרת החציון ברורה, אולם כאשר מספר הנתונים זוגי, ההגדרה אינה חד-משמעית.<br>

כך לדוגמה עבור הקבוצה 
</div>


$$\{1,2,5,10\}$$

<div dir='rtl'>
החציון יהיה
</div>

$$Me=\frac{2+5}{2}=3.5$$

נחזור לנתוני הגבהים והמשקלים, ונחשב את החציונים עבורם:
</div>

## אחוזונים (Quantiles)<br>
</div>

<div dir='rtl'>
החציון שלמדנו עליו, הוא מקרה פרטי של אחוזון, כי למעשה חציון הוא אחוזון 0.5.<br>
המשמעות של אחוזון היא נתון שמחלק את הנתונים, כך שהאחוז המבוקש מהנתונים בעל ערך נמוך מערכו של האחוזון.<br>
למעשה האחוזונים הם מדד למיקום יחסי, המאפשרים לנו לקבל תמונה רחבה על הטווחים המכילים את החלקים היחסיים השונים של הנתונים.<br>
לדוגמה, אחוזון 0.05 הוא הנתון ש-5% מהנתונים קטנים או שווים לו.<br>
לעיתים מסמנים אחוזון כך:
</div>

$$x_p$$

<div dir='rtl'>
כאשר p הוא אחוז הערכים הקטנים או שווים לאחוזון.<br>
עבור הדוגמה שלנו, נחשב אחוזונים בקפיצות של 0.05, ונקבל תמונה רחבה על הנתונים.
</div>

```python
quantiles = [0.05 * x for x in range(21)]
quantile_df = pd.DataFrame(index=["Height (CM)", "Weight (KG)"])

for q in quantiles:
    quantile_df[f"Quantile {q:.2f}:"] = Heights_and_Weights_df.quantile(q)

quantile_df.transpose()
```

<div dir='rtl'>
לדוגמה, 95%  מהמדגם בעלי משקל של עד 180.62 סנטימטרים.<br>
10% מהמדגם בעלי משקל של עד 50.85 ק"ג.<br>
יש הרבה מושגים הקשורים לאחוזונים, כמו רבעונים (אחוזוני 0.25, 0.5, 0.75) וטווח בין רבעוני. לא נרחיב על מושגים אלו.<br>
</div>

## שונות וסטיית תקן
</div>

<div dir='rtl'>
בקורס הסתברות למדנו על מדדי הפיזור שנקראים שונות (Variance) וסטיית תקן (Standard Deviation או std).<br>
למדנו גם על ממוצע הסטיות מהממוצע - MAE (Mean Absolute Error).<br>
הנוסחאות של המדדים הללו הן:
</div>

$$ Var(X) = \sum_{i=1}^{n} \frac{(x_i - \overline{x})^2}{n} $$
$$ std(X) = \sigma_X = \sqrt{Var(X)} $$
$$ MAE =  \sum_{i=1}^{n} \frac{|x_i - \overline{x}|}{n} $$

<div dir='rtl'>
בואו נעבור מהתאוריה ליישום.<br>
ניקח את הנתונים מהדוגמה שלנו, ושלב אחר שלב נוסיף עמודות עבור חישובי המדדים.<br>
אלה הנתונים הבסיסיים:
</div>

```python
Heights_and_Weights_df
```

<div dir='rtl'>
נוסיף עמודות עבור ההפרשים בין הנתונים לממוצעים.<br>
(תזכורת: הממוצעים של גובה ומשקל בהתאמה הם 172.7025 סנטימרים ו- 57.642 ק"ג)
</div>

```python
measures_of_variability_df = Heights_and_Weights_df.copy()
measures_of_variability_df["Height - Mean"] = measures_of_variability_df["Height (CM)"] - measures_of_variability_df["Height (CM)"].mean()
measures_of_variability_df["Weight - Mean"] = measures_of_variability_df["Weight (KG)"] - measures_of_variability_df["Weight (KG)"].mean()
measures_of_variability_df
```

<div dir='rtl'>
אז עמודות עבור ההפרשים בין הסטיות מהממוצעים (בערך מוחלט) ייראו כך:
</div>

```python
measures_of_variability_df["Height AE"] = abs(measures_of_variability_df["Height - Mean"])
measures_of_variability_df["Weight AE"] = abs(measures_of_variability_df["Weight - Mean"])

measures_of_variability_df[["Height AE","Weight AE"]]
```

<div dir='rtl'>
רואים בבירור שאלה הערכים שנוצרו קודם, רק בערך מוחלט.<br>
אז מדדי ה-MAE יהיו:
</div>

```python
MAE_Heights = measures_of_variability_df["Height AE"].mean()
MAE_Weights = measures_of_variability_df["Weight AE"].mean()
print(f"Heights MAE is: {MAE_Heights}.")
print(f"Weights MAE is: {MAE_Weights}.")
```

<div dir='rtl'>

אם ניתן לתוכנה לחשב את סטיית התקן, נקבל ערכים גבוהים במעט.<br>
ייתכן ונרחיב על כך בהמשך הקורס, אך לעת עתה אציין שהחישוב השונה נובע מרצון לתקן הטיות שיכולות להתרחש בעקבות צורת הדגימה.<br>
למעשה הנוסחה שבעזרתה מחשבות התוכנות את סטיית התקן היא:
</div>

$$ S_{n-1}^2 = \sum_{i=1}^{n} \frac{(x_i - \overline{x})^2}{n-1} $$

<div dir='rtl'>
ובמקרה שלנו הערכים שיתקבלו הם:
</div>

```python
SD_Heights = measures_of_variability_df["Height (CM)"].std()
SD_Weights = measures_of_variability_df["Weight (KG)"].std()
print(f"Heights SD is: {SD_Heights}.")
print(f"Weights SD is: {SD_Weights}.")
```

<div dir='rtl'>
למי שמתעניין איך להגיע לערכים שאנחנו חישבנו בעצמנו, אפשר להגביל את "דרגות החופש" כך:
</div>

```python
SD_Heights = measures_of_variability_df["Height (CM)"].std(ddof=0)
SD_Weights = measures_of_variability_df["Weight (KG)"].std(ddof=0)
print(f"Heights SD is: {SD_Heights}.")
print(f"Weights SD is: {SD_Weights}.")
```

<div dir='rtl'>
לסיום היחידה, אני רוצה להמחיש לכם את המשמעות של סטיית התקן.<br>
עבור התפלגות נורמלית (לא בדקנו שהנתונים שלנו מתפלגים נורמלית, נתייחס לזה בהמשך הקורס) 95% מערכים של מדגם צפויים להיות בטווח של עד
</div>

$$2 \cdot \sigma_X$$

 מהממוצע.<br>

</div>

$$ P(\mu -2 \cdot \sigma_X\leq X \leq \mu +2 \cdot \sigma_X) = 95\%$$
